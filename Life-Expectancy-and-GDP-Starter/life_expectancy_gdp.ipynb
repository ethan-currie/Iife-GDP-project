## Life Expectancy vs GDP Project


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Loading data into dataframe, and exploring the data
data = pd.read_csv('all_data.csv')
print(data.head())
print(data.Country.unique())
print(data.Year.unique())
```

We can see from the exploration that there are 6 countries present in the dataset : Chile, China, Germany, Mexico, USA and Zimbabwe. Also the data ranges from 2000 to 2015

```python
#Chile

chile = data[data.Country == 'Chile']

mean_life_expectancy_chile = round(chile['Life expectancy at birth (years)'].mean(), 1)
print(f"The mean life expectancy in years for people from Chile between 2000 and 2015 is {mean_life_expectancy_chile} years \n")

mean_gdp_chile = round(chile['GDP'].mean() / 10**12, 4)
print(f"The mean GDP of Chile between 2000 and 2015 is ${mean_gdp_chile} trillion \n")

fig, axes = plt.subplots(1,3, figsize = (20,5))

sns.regplot(ax = axes[0], x = 'Year', y = 'Life expectancy at birth (years)', data = chile, fit_reg = False)
axes[0].set_title("How life expectancy has changed over time in Chile")



sns.regplot(ax = axes[1], x = 'Year', y = 'GDP', data = chile, fit_reg = False)
axes[1].set_title("How GDP has changed over time in Chile")


sns.regplot(ax = axes[2], y = 'GDP', x = 'Life expectancy at birth (years)', data = chile, fit_reg = False)
plt.title("Life expectancy vs GDP in Chile")

plt.show()
plt.clf()
```

From the plots we can see:

-  Both the life expectancy and GDP of Chile has grown overtime
-  There is a positive correlation between the GDP and life expectancy
